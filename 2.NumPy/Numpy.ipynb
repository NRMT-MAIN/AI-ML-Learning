### NumPy
**(Numerical Python)**

- Used for working with arrays.
- In Python we have lists that serve the purpose of arrays, but they are slow to process.
- NumPy is incredible library to perform mathematical and statistical operations due to it's fast and memory efficient as it is optimized to work with latest CPU architectures.
- NumPy aims to provide an array object that is up to 50x faster than traditional Python lists. So processes can access and manipulate them very efficiently.

##### Functions for creating 1D array

arange
- arange is an array-valued version of the built-in Python range function
- **Syntax:** `np.arange(start, end, step)`

```python
a = np.arange(10)
print(a)
```

linspace
- Return evenly spaced numbers over a specified interval.
- **Syntax:** `np.linspace(start, end, num_of_samples)`

```python
np.linspace(1, 5, 5)
```

zeros
- Generate arrays of zeros
- **Syntax:** `np.zeros(shape)`

```python
np.zeros((2,))
```

ones
- Generate arrays of ones
- **Syntax:** `np.ones(shape)`

```python
np.ones((2,))
```

random.randint
- Return random integers from start (inclusive) to end (exclusive)
- **Syntax:** `np.random.randint(start, end, num_of_samples)`

```python
np.random.randint(1, 1000, 5)
```

##### Accessing elements / Extracting the values
1. Accessing a single value through Indexing
2. Accessing multiple values based on slicing
3. Accessing multiple values based on indexing
4. Accessing multiple values based on condition

```python
a = np.arange(10, 51, 10)
print(a)
```

1. Accessing a single value through Indexing

```python
# positive indexing
a[1]
```

```python
# negative indexing
a[-1]
```

2. Accessing multiple values based on slicing

```python
a[1:5:2]
```

```python
a[-1:-4:-1]
```

3. Accessing multiple values based on indexing

Syntax: `array[list of index numbers]`

```python
a[[0, 3, 2, 4]]
```

4. Accessing multiple values based on condition (Mask Indexing or Fancy Indexing)

Syntax: `array[condition]`

-- first, it will check the condition; the item which satisfies the condition (means which item gives True for the condition), that item will be selected

```python
a = np.array([45, 46, 47])
a[a % 2 == 0]
```

##### Modify the values in Numpy

Add a single value in an array
Syntax: `np.append(array, value)`

```python
a = np.array([45, 46, 47])
a1 = np.append(a, 110)
print(a1)
```

Add multiple values in an array

```python
a = np.array([45, 46, 47])
a2 = np.append(a, [140, 120, 130])
print(a2)
```

Replacing a single value in a 1D array

```python
a[0] = 150
print(a)
```

Replacing multiple values with a single value in an array

```python
a[[1, 2]] = 15
a
```

Replacing multiple values with multiple values in an array

```python
a[[1, 2]] = [460, 14]
a
```

Delete a single value in an array
Syntax: `np.delete(array, indexnumber)`

```python
a3 = np.delete(a, 1)
print(a3)
```

Delete multiple values in an array
Syntax: `np.delete(array, [multiple index numbers])`

```python
a4 = np.delete(a, [1, 2])  # deleting multiple values
print(a4)
```

##### Sort
Sorting along an axis: `axis=1` --> row and `axis=0` --> column

```python
# Column-wise sorting
a = np.array([[5, 4, 6], [2, 8, 2], [8, 9, 20]])
b = np.sort(a, axis=0)
b
```

```python
# Row-wise sorting
b = np.sort(a, axis=1)
b
```

```python
a = np.array([[1, 2], [3, 4], [5, 6]])
b = np.array([[40, 4], [14, 24], [34, 44]])

# adding a and b
print(a + b)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# dot product (multiply a and b - elementwise multiplication)
print(a * b)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
print(a==3)
print(a==b)
```

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

# matrix multiplication
np.matmul(a, b)
```

```python
## Combinning arrays
np.concatenate((a, b), axis=0)
```

```python
# vertical stack (equal to np.concatenate((a,b), axis=0))
np.vstack((a, b))
```

```python
np.concatenate((a, b), axis=1)
```

```python
# horizontal stack (equal to np.concatenate((a,b), axis=1))
np.hstack((a, b))
```
