Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. In simpler terms, søntax is about the correct arrangement of words and symbols in a code.

Semantics refers to the meaning or the interpretation of the symbols, characters, and commands in a language. It is about what the code
is supposed to do when it runs.

```python
## Basic Syntax Rules in python
## Case Sensitivity - Python is case sensitive

name="Nirmit"
Name="Sahu"
print(name)
print(Name)
```

#### Indentation
Indentation in Python is used to define the structure and hierarchy of the code. Unlike many other programming languages that use braces {} to delimit blocks of code, Python uses indentation to determine the grouping of statements. This means that all the statement within a block must be indented at the same level.

```python
## Indentation - Python uses indentation to define blocks of code. Indentation is typically 4 spaces.
if name == "Nirmit":
    print("Hello, Nirmit!")
print("This is outside the if block.")
```

```python
## This is single line comment
""" this is
multiple line comment
"""
```

```python
## Line Continuation - Python allows you to break long lines of code into multiple lines for better readability. 
# You can use a backslash (\) to indicate that a line continues on the next line.

total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)
```

```python
# Multiple statements on a single line - You can use a semicolon (;) to separate multiple statements on a single line, 
# but it's generally not recommended for readability.
x = 10; y = 20; z = x + y
print(z)
```

```python
## Understanding semantics in python
# Semantics refers to the meaning of the code, while syntax refers to the structure of the code.

age=25 # kernel automatically understands that age is an integer
name="Nirmit" # kernel automatically understands that name is a string
print(type(age)) # <class 'int'>
print(type(name)) # <class 'str'>

#Type inference - Python uses dynamic typing, which means that the type of a variable is determined at runtime based on the value assigned to it.
x = 10 # x is an integer
x = "Hello" # x is now a string
print(type(x)) # <class 'str'>
```

```python
## Code example of indentation and semantics
if True:
    print("This is indented correctly.")
    if False:
        print("This will not be printed.")
    print("This is still part of the first if block.")
print("This is outside the if block.")
```
