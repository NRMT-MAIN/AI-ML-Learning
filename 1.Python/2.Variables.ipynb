#### Variables
Variables are the fundamental elements in programming used to store data that can be referenced and manipulated in a program.

```python
a=1000
print(a)
```

```python
# Declaring and Assigning a variable
age = 25
height = 175.5
name = "Alice"
is_student = True

# Printing the variables
print("Age:", age)  
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

```python
# Naming conventions
# 1. Use descriptive names
# 2. Use snake_case for variable names
# 3. Use PascalCase for class names
# 4. Use UPPERCASE for constants
# Example of constants
PI = 3.14159
SPEED_OF_LIGHT = 299792458  # in meters per second
print("Value of PI:", PI)
print("Speed of Light:", SPEED_OF_LIGHT)
#2name is invalid
#first-name is invalid

```

```python
# Understanding variable types
## In Python, variables can hold different types of data. 
# The type of a variable is determined by the value assigned to it. 
# Here are some common variable types in Python:

age = 30  # Integer
height = 180.5  # Float
name = "Bob"  # String
is_student = False  # Boolean
print("Age:", age, "Type:", type(age))
print("Height:", height, "Type:", type(height))
print("Name:", name, "Type:", type(name))
```

```python
# Type checking and type conversion
# You can check the type of a variable using the type() function.
# You can also convert between types using functions like int(), float(), str(), etc.
age_str = str(age)  # Convert integer to string
height_int = int(height)  # Convert float to integer
print("Age as string:", age_str, "Type:", type(age_str))
print("Height as integer:", height_int, "Type:", type(height_int))
print("Is Student as string:", str(is_student), "Type:", type(str(is_student)))

# print(int(name))  # This will raise an error because 'name' cannot be converted to an integer
```

```python
## Dynamic typing
# In Python, you can reassign a variable to a different type of value.
my_variable = 10  # Initially an integer
print("Value:", my_variable, "Type:", type(my_variable))
my_variable = "Now I'm a string"  # Now it's a string
print("Value:", my_variable, "Type:", type(my_variable))
```

```python
## input - By default, the input() function returns a string. You can convert it to other types if needed.
age=int(input("Enter your age: "))
print("You entered:", age, "Type:", type(age))
```
