### Pandas Series

- It is a one-dimensional (1D) array holding data of any type.
- A Pandas Series is like a column in a table.

> **Note:**
> 1. dtype is important
> 2. Unlike NumPy, data will not be converted according to maximum data type.
> 3. Unlike NumPy, error will not be thrown when in heterogenous shape
>
> `ndim`, `shape` etc. are same as NumPy
>
> In Numpy, only list is used to convert to array. In pandas, both list and dictionary are used to convert to series.
>
> In pandas we can provide index other than number — understand by `iloc` and `loc`.
>
> For patterns like arange, linspace you can use numpy and then convert to series.

Accessing element in the series

```python
d = {'a': 10, 'b': 20,'c': 30}

S = pd.Series(d)
print(S)
print(S['a'])
print(S.iloc[2])
print(S.loc['c'])
```

```python
# Accesing the multiple values based on slicing
a = [15, 19, 23, 24, 37, 42, 46]
s = pd.Series(a , index=["a", "b", "c", "d", "e", "f", "g"])
s.loc["a" : "d"]
```

```python
# Accesing the multiple values based on indexing
s.iloc[[1, 0, 2, 3]]
```

```python
#print(s.loc[0]) # This will give an error because the index is not integer based
```

```python
# Keys can be repeated in a Series
a = [15, 19, 23, 24, 37, 42, 46]
s = pd.Series(a , index=["a", "b", "c", "d", "e", "f", "a"])
print(s)
# Accessing the value of the repeated key
print(s.loc["a"])
print(s.iloc[0]) # This will give the first occurrence of the repeated key
```

```python
#extract based on single condition
print(s[s > 30])
```

##### Modification / Data Manipulation

```python
# replace a single value with single value
s.loc["a"] = 100
print(s)
```

```python
# replace multiple values with single value
s.loc[["b", "c", "d"]] = 50
print(s)
s.iloc[[1, 2, 3]] = 5
print(s)
```

```python
# Add a new value into the Series
s.loc["h"] = 55
print(s)
```

```python
# remove a value from the Series
s.drop("h", inplace=True)
print(s)
```

```python
# remove multiple values from the Series
s.drop(index=["a", "b", "c"])
print(s)
s=s.drop(index=["a", "b", "c"])
print(s)
```

```python
s1 = pd.Series([10,20,30])
s2 = pd.Series([1,2,3,4])
print(s1+s2) 
# The addition is performed element-wise, and since s2 has an extra element, it will be ignored in the addition.
# Unlike NumPy arrays, Pandas Series can handle operations between Series of different lengths without throwing an error. 
# The resulting Series will have the same length as the longer Series, and the missing values will be filled with NaN (Not a Number) for the shorter Series.
```

##### 1. Create DataFrame using Dictionary

```python
## In Series - keys are the index/row here and values are storeed row-wise
## In DataFrame - keys are the column here and values are stored column-wise

d = {'col1': [1, 2], 'col2': [3, 4]}
s = pd.Series(d)
print(s)
```

# Combining DataFrames

```python
df1 = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore", "hyderabad"],
    "temperature": [32, 45, 40, 36]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["delhi", "mumbai", "banglore", "chennai"],
    "humidity": [68, 65, 75, 70]
})
df2
```

### Concatenation

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

### Merging of DataFrame

```python
# Left Join
left_df = pd.merge(df1, df2, on='city', how='left')
left_df
```

```python
# Right Join
right_df = pd.merge(df1, df2, on='city', how='right')
right_df
```

```python
# Inner Join -- common data records
inner_df = pd.merge(df1, df2, on='city', how='inner')
inner_df
```

```python
# Outer Join - all the records of both dataframes
outer_df = pd.merge(df1, df2, on='city', how='outer')
outer_df
```
