### Pandas Series

- It is a one-dimensional (1D) array holding data of any type.
- A Pandas Series is like a column in a table.

> **Note:**
> 1. dtype is important
> 2. Unlike NumPy, data will not be converted according to maximum data type.
> 3. Unlike NumPy, error will not be thrown when in heterogenous shape
>
> `ndim`, `shape` etc. are same as NumPy
>
> In Numpy, only list is used to convert to array. In pandas, both list and dictionary are used to convert to series.
>
> In pandas we can provide index other than number — understand by `iloc` and `loc`.
>
> For patterns like arange, linspace you can use numpy and then convert to series.

#### 1. Pandas Series can be created using List

```python
l = [10, 20, 30]
s = pd.Series(l)
print(s)
```

```python
type(s)
```

```python
s.shape
```

```python
# Series with custom index
l = [10, 20, 30]
s = pd.Series(l, index=["a", "b", "c"])
print(s)
```

```python
# Series from a list with float values
a = [1, 2, 3.2]
arr_1 = pd.Series(a)
print(arr_1)
```

```python
a = [True , 20 , 21.5 , "srk"]
arr_1 = pd.Series(a)
print(arr_1)
```

#### 2. Create Series using 1D Array

```python
a = np.array([10, 20, 30])
s = pd.Series(a)
print(s)
```

```python
# 1D array with custom index
a = np.array([10, 20, 30])
s = pd.Series(a, index=["a", "b", "c"])
print(s)
```

```python
pd.Series(np.arange(5,51,6))
```

```python
pd.Series(np.random.randint(1,100,10))
```

```python
pd.Series(np.zeros(5,dtype=int) + 5)
```

#### 3. Create Series using Dictionary

```python
d = {'a': 10, 'b': 20, 'c': 30}
s = pd.Series(d)
print(s)
```

#### Accessing element in the series

```python
d = {'a': 10, 'b': 20,'c': 30}

S = pd.Series(d)
print(S)
print(S['a'])
print(S.iloc[2])
print(S.loc['c'])
```

```python
# Accesing the multiple values based on slicing
a = [15, 19, 23, 24, 37, 42, 46]
s = pd.Series(a , index=["a", "b", "c", "d", "e", "f", "g"])
s.loc["a" : "d"]
```

```python
# Accesing the multiple values based on indexing
s.iloc[[1, 0, 2, 3]]
```

# Pandas DataFrame

- A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.

## 1. Create DataFrame using Dictionary

```python
d = {'col1': [1, 2], 'col2': [3, 4]}
s = pd.Series(d)
print(s)
```

```python
d = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(d)
df
```

```python
type(df)
```

```python
df.shape
```

## 2. Create DataFrame using Nested List

```python
data = [[1, 2, 3], [4, 5, 6]]
df1 = pd.DataFrame(data, columns=["a", "b", "c"])
df1
```

## 3. Create DataFrame using 2D array

```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
df2 = pd.DataFrame(arr_2d, index=["a", "b"], columns=["s", "r", "k"])
df2
```

# Pandas

Working with a sample DataFrame

```python
df = pd.DataFrame({
    'Age': [22, 30, 23, 25, 24],
    'Salary': [28000, 17000, 46000, 42000, 55000],
    'Gender': ["M", "F", "M", "M", "F"]
})
df
```

### set index

```python
df1 = df.set_index("Age")
df1
```

### reset index

```python
df1.reset_index()
```

### Drop a single row

```python
df1 = df.drop(index=[3])
df1
```

### Drop multiple rows

```python
df1 = df.drop(index=[0, 1, 2])
df1
```

### Drop a single column

```python
df1 = df.drop(columns=["Age"])
df1
```

### Drop multiple columns

```python
df1 = df.drop(columns=["Age", "Gender"])
df1
```

### Sorting values in a dataframe

```python
# Sorting in ascending order
df.sort_values(by='Age', ascending=True)
```

```python
# Sorting in descending order
df.sort_values(by='Age', ascending=False)
```

# Combining DataFrames

```python
df1 = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore", "hyderabad"],
    "temperature": [32, 45, 40, 36]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["delhi", "mumbai", "banglore", "chennai"],
    "humidity": [68, 65, 75, 70]
})
df2
```

### Concatenation

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

### Merging of DataFrame

```python
# Left Join
left_df = pd.merge(df1, df2, on='city', how='left')
left_df
```

```python
# Right Join
right_df = pd.merge(df1, df2, on='city', how='right')
right_df
```

```python
# Inner Join -- common data records
inner_df = pd.merge(df1, df2, on='city', how='inner')
inner_df
```

```python
# Outer Join - all the records of both dataframes
outer_df = pd.merge(df1, df2, on='city', how='outer')
outer_df
```
